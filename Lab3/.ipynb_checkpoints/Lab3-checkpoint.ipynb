```python
training_data = FashionMNIST(
    root="./data",
    train=True,
    download=True,
)

test_data = FashionMNIST(
    root="./data",
    train=False,
    download=True,
)
```

```python
def show_transformed_images(dataset):
    #     torch.manual_seed(1)
    torch.manual_seed(3)
    loader = DataLoader(dataset, batch_size=6, shuffle=True)
    batch = next(iter(loader))
    images, labels = batch

    grid = torchvision.utils.make_grid(images, nrow=3)
    plt.figure(figsize=(11, 11))
    plt.imshow(np.transpose(grid, (1, 2, 0)))
```

```python
class SiamenseNet(nn.Module):
    
    def __init__(self, feature_learning_net):
        super().__init__()
        
        self.feature_learning = feature_learning_net
        
        self.similarity = nn.Sequential(
            torch.nn.Linear(in_features=2000, out_features=100),
            torch.nn.ReLU(),
            torch.nn.Linear(in_features=100, out_features=10),
            torch.nn.ReLU(),
            torch.nn.Linear(in_features=10, out_features=1),
            torch.nn.Sigmoid())

        
    def forward(self, data):

        features_extracted_img1 = self.feature_learning(data[0])
        features_extracted_img2 = self.feature_learning(data[1])
        
        flatten_f_map = torch.hstack((features_extracted_img1, features_extracted_img2))
        similarity = self.similarity(flatten_f_map)
        return similarity
```

```python
class MyData(Dataset):

    def __init__(self, dataset, preprocess):
        self.y = dataset.targets
        self.X = dataset.data
        self.groups = np.array([self.X[np.where(self.y == i)[0]] for i in np.unique(self.y)])
        self.group_lengths = np.array([len(group) for group in self.groups])
        self.match_group = True
        self.transforms = preprocess
        self.all_groups = range(len(self.groups))

        
    def __getitem__(self, index):
        
        out_img_a, out_img_b, out_score = None, None, None
        
        group_idx = np.random.choice(self.all_groups)
        
        out_img_a = self.groups[group_idx][np.random.choice(self.group_lengths[group_idx])]
        out_score = int(self.match_group)
        
        if self.match_group:
            out_img_b = self.groups[group_idx][np.random.choice(self.group_lengths[group_idx])]
        else:
            group_idx = np.random.choice([group for group in self.all_groups if group != group_idx])
            out_img_b = self.groups[group_idx][np.random.choice(self.group_lengths[group_idx])]
        
        self.match_group = not self.match_group
        return self.transforms(out_img_a), self.transforms(out_img_b), out_score

    def __len__(self):
        return len(self.y)
```

```python
preprocess = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize(48),
#     transforms.CenterCrop(48),
    transforms.Grayscale(num_output_channels=3),
    transforms.ToTensor(),
#     transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
my_training_data, my_test_data = MyData(training_data, preprocess), MyData(test_data, preprocess)
```

```python
train_loader, test_loader = DataLoader(my_training_data, batch_size=32), DataLoader(my_test_data, batch_size=1024)
```

```python
train_sample = next(iter(DataLoader(my_training_data, batch_size=6)))
pv_a, pv_b, pv_sim = train_sample
fig, m_axs = plt.subplots(2, pv_a.shape[0], figsize = (15, 6))
for c_a, c_b, c_d, (ax1, ax2) in zip(pv_a, pv_b, pv_sim, m_axs.T):
    ax1.imshow(c_a.permute(1, 2, 0))
    ax1.set_title('Image A')
    ax1.axis('off')
    ax2.imshow(c_b.permute(1, 2, 0))
    ax2.set_title('Image B\n Similarity: %3.0f%%' % (100*c_d))
    ax2.axis('off')
```

```python
def show_model_output(similarity_model, batch_size = 6):
    pv_a, pv_b, pv_sim = next(iter(DataLoader(my_training_data, batch_size = batch_size)))
    # a completely untrained model
    pred_sim = similarity_model([pv_a, pv_b]);
    fig, m_axs = plt.subplots(2, pv_a.shape[0], figsize = (15, 6));
    for c_a, c_b, c_d, p_d, (ax1, ax2) in zip(pv_a, pv_b, pv_sim, pred_sim, m_axs.T):
        ax1.imshow(c_a.permute(1, 2, 0));
        ax1.set_title('Image A\n Similarity: %3.0f%%' % (100*c_d));
        ax1.axis('off');
        ax2.imshow(c_b.permute(1, 2, 0));
        ax2.set_title('Image B\n Similarity: %3.0f%%' % (100*p_d));
        ax2.axis('off');

_ = show_model_output(SiamenseNet(resnext50_32x4d).to('cpu')) 
```

```python
def train(dataloader, model, loss_fn, optimizer):
#     model.train()
    avg_loss=0
    for i,(img_a, img_b, labels) in enumerate(dataloader):
        optimizer.zero_grad()
        if torch.cuda.is_available():
            img_a = Variable(img_a.cuda())
            img_b = Variable(img_b.cuda())
            labels = Variable(labels.cuda())
        
        pred = model([img_a, img_b])
        loss = loss_fn(pred, labels.reshape(pred.shape).float())
        
        avg_loss += loss
        
        loss.backward()
        optimizer.step()
    avg_loss /= (i+1)
    return avg_loss
```

```python
for param in resnext50_32x4d.parameters():
    param.requires_grad = False
```

```python
resnext50_32x4d=resnext50_32x4d.to(device)
```

```python
model = SiamenseNet(resnext50_32x4d)
```

```python
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(),lr=0.001)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using {} device".format(device))
```

```python
train(train_loader, model, criterion, optimizer)
```
