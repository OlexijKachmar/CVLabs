```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Image loading

```python
def get_rgb_matrix(img_path:str):
    img_bgr = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    return img_rgb
```

```python
high_contrast_img = get_rgb_matrix(img_path='./HighContrastImage.png')
low_contrast_img = get_rgb_matrix(img_path='./LowContrastImage.png')

img_higher_lighting = get_rgb_matrix(img_path='./cup_high_int.png')
img_lowerer_lighting = get_rgb_matrix(img_path='./cup_low_int.png')
```

# Distortion

```python
down_sampled_img = cv2.pyrDown(cv2.pyrDown(high_contrast_img))
num_rows, num_cols = down_sampled_img.shape[:2]
```

```python
rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), 30, 1)
distorted_image = cv2.warpAffine(down_sampled_img, rotation_matrix, (num_cols, num_rows))
```

```python
fx, plots = plt.subplots(1, 2, figsize=(20,10))

plots[0].imshow(high_contrast_img)
plots[1].imshow(distorted_image)
```

# Extracting features

```python
from collections import namedtuple
```

```python
image_features = namedtuple('image_features', ['keypoints_1', 'descriptors_1', 'keypoints_2', 'descriptors_2'])
```

```python
def extract_visualize_keypoints(initial_img, different_img):

    sift = cv2.xfeatures2d.SIFT_create()
    gray_base = cv2.cvtColor(initial_img, cv2.COLOR_BGR2GRAY)
    gray_distorted = cv2.cvtColor(different_img, cv2.COLOR_BGR2GRAY)
    
    keypoints_1, descriptors_1 = sift.detectAndCompute(gray_base, None)
    keypoints_2, descriptors_2 = sift.detectAndCompute(gray_distorted, None)
    
    keypoints_clean_img = np.copy(initial_img)
    keypoints_distorted_img = np.copy(different_img)
    
    cv2.drawKeypoints(initial_img, keypoints_1, keypoints_clean_img, color = (0, 255, 0))
    
    cv2.drawKeypoints(different_img, keypoints_2, keypoints_distorted_img, color = (0, 255, 0))
    
    # Display image with and without keypoints size
    fx, plots = plt.subplots(1, 2, figsize=(20,10))
    
    plots[0].set_title("Initial image keypoints")
    plots[0].imshow(keypoints_clean_img, cmap='gray')
    
    plots[1].set_title("Different image keypoints")
    plots[1].imshow(keypoints_distorted_img, cmap='gray')
    
    print("Number of Keypoints Detected In The Initial Image: ", len(keypoints_1))
    
    print("Number of Keypoints Detected In The Different Image: ", len(keypoints_2))
    
    return image_features(keypoints_1, descriptors_1, keypoints_2, descriptors_2)
```

```python
high_contrast_features = extract_visualize_keypoints(high_contrast_img, distorted_image)
```

```python
different_lighting_features = extract_visualize_keypoints(img_higher_lighting, img_lowerer_lighting)
```

# Descriptors matching

```python
def draw_matches(satisfactory_matches, **drawing_kwargs):
        
    sorted_matches = sorted(satisfactory_matches, key = lambda x:x.distance)[:10]
    
    drawing_kwargs['matches1to2'] = sorted_matches
    matched_images = cv2.drawMatches(**drawing_kwargs)  
    
    plt.rcParams['figure.figsize'] = [15.0, 10.0]
    plt.title('Best Matching Points')
    plt.imshow(matched_images)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck = False)
satisfactory_matches = bf.match(high_contrast_features.descriptors_1, high_contrast_features.descriptors_2)
draw_matches(satisfactory_matches, img1 = high_contrast_img, keypoints1=high_contrast_features.keypoints_1,\
             img2 = distorted_image, keypoints2 = high_contrast_features.keypoints_2,outImg=None,
             flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck = False)
satisfactory_matches = bf.match(different_lighting_features.descriptors_1, different_lighting_features.descriptors_2)
draw_matches(satisfactory_matches, img1 = img_higher_lighting, keypoints1= different_lighting_features.keypoints_1,\
             img2 = img_lowerer_lighting, keypoints2 = different_lighting_features.keypoints_2,outImg=None,
             flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

```python
def knn_matcher(descriptor_1, descriptor_2, threshold = 0.15):
    
    top = 2
    perfect_matches = []
    for query_idx in range(descriptor_1.shape[0]):
        
        distances = np.linalg.norm(descriptor_1[query_idx] - descriptor_2, axis=1)
        
        top_matched_descriptors = np.argsort(distances)[:top]
        top_distances = sorted(distances)[:top]
        
        dmatch_obj = [cv2.DMatch(_queryIdx=query_idx,\
                                 _trainIdx=train_idx,\
                                 _distance=distance) \
                     for train_idx, distance in zip(top_matched_descriptors, top_distances)]
        
        if dmatch_obj[0].distance < threshold*dmatch_obj[1].distance:
            perfect_matches.append(dmatch_obj[0])

    return perfect_matches
```

```python
perfect_matches = knn_matcher(different_lighting_features.descriptors_1, different_lighting_features.descriptors_2,\
                              threshold=0.6)
```

```python
draw_matches(perfect_matches, img1 = img_higher_lighting, keypoints1=different_lighting_features.keypoints_1,\
             img2 = img_lowerer_lighting, keypoints2 = different_lighting_features.keypoints_2,outImg=None,
             flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

```python
perfect_matches = knn_matcher(high_contrast_features.descriptors_1, high_contrast_features.descriptors_2,\
                              threshold=0.15)
```

```python
draw_matches(perfect_matches, img1 = high_contrast_img, keypoints1 = high_contrast_features.keypoints_1,\
             img2 = distorted_image, keypoints2 = high_contrast_features.keypoints_2,outImg=None,
             flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```
